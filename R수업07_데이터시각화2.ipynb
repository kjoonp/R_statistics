```
# 히스토그램
# 연속형 데이터를 도수분포표를 이용해서 막대그래프로 나타낸것
# 도수분포표 : 특정 구간에 속하는 자료의 갯수를 나타낸표
# x축은 변수의 구간을 의미하는계급
# y 축은 구간내 빈도를 의미하는 도수
# hist(대산, 구간정보)
```

```
# 도수분포표를 만들려면 먼저 데이터를 크기순으로 정렬 / 그런다음 구간별로 빈도를 계산해서 시각화
```

```
hist(heights, main = '학생키 빈도')
h <- hist(heights, main = '학생키 빈도')
h
# $breaks : 구간정보, 보통 5,10,15 ..배수로 나눔
# $counts : 구간별 도수 (빈도수)
# $density : 확률밀도 추정값 (해당구간에서 각도수가 나타날 확률) 
# $mids : 구간의 중간값
# $equidist : 구간의 균등분할 여부
```

```
# 구간을 임의로 설정하려면 breaks 속성 사용
hist(heights, breaks=10)
```

```
# 히스토그램에 수치 나타내기
# text (x축위치, y축위치, 표시할 값)
h <- hist(heights, breaks=10, ylim=c(0,3))
text(h$mids, h$counts+0.1, cex=1.5, labels=h$counts)
```

```
# 상대도수 곡선 히스토그램
# y축이 density로 바뀜 : 확률밀도 추정값
hist(heights, probability = T, xlim=c(120,200))
lines(density(heights), col='red')
```
